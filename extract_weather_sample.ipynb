```python
import requests
import pandas as pd

cities = { "city": { "name": "Lagos", "coordinates": { "latitude": 6.6137, "longitude": 3.3553 } } }
start_date = "2023-01-01"
end_date =  "2023-12-31"

def get_daily_weather(cities, date1, date2):
    url = f"https://archive-api.open-meteo.com/v1/archive"

    params = {
        "latitude": cities["city"]["coordinates"]["latitude"],
        "longitude": cities["city"]["coordinates"]["longitude"],
        "start_date": date1,
        "end_date": date2,
        "daily": ["temperature_2m_max", "precipitation_sum", "wind_speed_10m_max", ]
    }
    response = requests.get(url, params=params)
    weather_data = response.json()
    
    weather_data_rows = pd.DataFrame(weather_data["daily"])
    weather_data_rows["zone"] = "North West"
    print("Columns before save:", weather_data_rows.columns.tolist())
    weather_data_rows.to_csv("weather_data_sample.csv", index=False)
    
    print(f"Saved {len(weather_data_rows):,} rows to 'weather_data_sample.csv'")

get_daily_weather(cities, start_date, end_date)

    
```
